>  * Next, we want to generate a slightly more complicated graphic. I'm interested to know if there are certain crimes that happen much more in certain neighborhoods than what's typical. Below I describe how to get that plot going:
>    - First, we need to calculate the relative probabilities of seeing each type of crime in the dataset as a whole. Let's call it `P(crime)`.
>    - Next, we calculate that same probability distribution _but for each PD district_, let's call that `P(crime|district)`.
>    - Now we look at the ratio `P(crime|district)/P(crime)`. That ratio is equal to 1 if the crime occurs at the same level within a district as in the city as a whole. If it's greater than one, it means that the crime occurs _more frequently_ within that district. If it's smaller than one, it means that the crime is _rarer within the district in question_ than in the city as a whole.

````python
df['Incident Year'] = pd.to_datetime(df['Incident Datetime']).dt.year
df_early = df[df['Incident Year'] <= 2008]
df_new = df[df['Incident Year'] > 2020]
````

````python
# Create a frequency table (counts)
def calc_p_crime_given_pd(df):
    counts = pd.crosstab(df['Incident Category'], df['Police District'])

    # Add a 'Total' column for the whole city's crime counts
    counts['City_Total'] = counts.sum(axis=1)

    counts

    # We divide each column by its own sum
    p_crime_given_district = counts.divide(counts.sum(axis=0), axis=1)

    # Calculate P(crime) - the global probability
    p_crime = counts['City_Total'] / counts['City_Total'].sum()

    # Calculate the ratio: P(crime|district) / P(crime)
    # We divide the district-specific probabilities by the global vector
    ratios = p_crime_given_district.divide(p_crime, axis=0)

    # Drop the 'City_Total' column from ratios as it will just be 1.0
    ratios = ratios.drop(columns=['City_Total'])
    return ratios

ratios_early = calc_p_crime_given_pd(df_early)
ratios_new = calc_p_crime_given_pd(df_new)
````

````python
# 1. Identify every possible Crime and every possible District across both eras
all_crimes = sorted(list(set(ratios_early.index) | set(ratios_new.index)))
all_pds = sorted(list(set(ratios_early.columns) | set(ratios_new.columns)))

# 2. Force both DataFrames to use this "Master Grid"
# Any missing crime in a district will be filled with 0.0
r_early_final = ratios_early.reindex(index=all_crimes, columns=all_pds, fill_value=0)
r_new_final = ratios_new.reindex(index=all_crimes, columns=all_pds, fill_value=0)
````

````python
from scipy import stats
# Calculate mean for x, y
# format so only two decimals

fig, ax = plt.subplots(2,2)
ax_flat = ax.flatten()

for i,data in enumerate([data1, data2, data3, data4]):
    x = data[:,0]
    y = data[:,1]
    
    a, b, r_value, p_value, std_err = stats.linregress(x,y)
    print(f'Dataset {i+1}:')
    print(f'    X: {format(np.mean(x), ".2f" )} - Y:{format(np.mean(y), ".2f")}')
    print(f'    Var X: {format(np.var(x), ".3f" )} - Var Y:{format(np.var(y), ".3f")}')
    print(f'    Pearson Corr: {[format(i, ".3f") + " " + format(j, ".3f") for i, j in np.corrcoef(x, y)]}')
    print(f'    Linear Regression: {[format(i, ".3f") for i in [a, b, r_value, p_value, std_err]]}')
    
    ax_flat[i].plot(x, y, marker='o', linestyle='none')
    x_line = np.array([x.min(), x.max()])
    y_line = a * x_line + b
    
    ax_flat[i].plot(x_line, y_line, 'r-', label=f'y={a:.2f}x+{b:.2f}')

````

````python
# Jitter plot
# crime type and a time interval
# on the hour 13-14
import random 

crime = 'BURGLARY'
start_date = '2018-01-01'
end_date = '2018-07-01'
hour = '13'

mask = (
    (df['Incident Category'] == crime)
    & (df['Incident Date'] >= start_date)
    & (df['Incident Date'] < end_date)
    & (df['Incident Time'].fillna('').str[:2] == hour)
)

df_filtered = df.loc[mask]

x = df_filtered['Incident Time'].str[3:].astype('int')
y = np.random.normal(loc=0, scale=0.05, size=len(x))

    
# Count no of instances
fig, ax = plt.subplots()
ax.plot(x, y, marker ='o', linestyle='none',  markerfacecolor='none', alpha=0.5)
ax.yaxis.set_visible(False)
ax.set_xlabel('Minute within hour (13:00–13:59)')
ax.set_title('Burglary incidents between 13:00–14:00 during 1st january to 1st july 2018')

plt.show()
````

````python
# Histogram
# Two crimes,

crimes = ['BURGLARY', 'LOST PROPERTY']
start_date = '2018-01-01'
end_date = '2018-07-01'


mask = (
    (df['Incident Category'].isin(crimes))
    & (df['Incident Date'] >= start_date)
    & (df['Incident Date'] < end_date)
)

df_filtered = df.loc[mask]

# Ensure numeric
df_filtered['Latitude'] = pd.to_numeric(df_filtered['Latitude'], errors='coerce')
df_filtered['Longitude'] = pd.to_numeric(df_filtered['Longitude'], errors='coerce')

fig, axes = plt.subplots(2, 2, figsize=(12, 8))

for row, crime in enumerate(crimes):
    df_c = df_filtered[df_filtered['Incident Category'] == crime]

    lat = df_c['Latitude'].dropna()
    lon = df_c['Longitude'].dropna()

    # Latitude histogram
    axes[row, 0].hist(lat, bins=50, edgecolor='black')
    axes[row, 0].set_title(f'{crime} – Latitude')
    axes[row, 0].set_xlabel('Latitude')
    axes[row, 0].set_ylabel('Count')

    # Longitude histogram
    axes[row, 1].hist(lon, bins=50, edgecolor='black')
    axes[row, 1].set_title(f'{crime} – Longitude')
    axes[row, 1].set_xlabel('Longitude')
    axes[row, 1].set_ylabel('Count')

plt.tight_layout()
plt.show()
````

````python
# KDEs
from scipy.stats import gaussian_kde

fig, axes = plt.subplots(2,1,sharex=True)

# Extract minute-of-hour
minutes = (
    df_filtered['Incident Time']
    .str[3:]
    .astype(int)
    .values
)

# KDE on full data
x_grid = np.linspace(0, 59, 500)

kde = gaussian_kde(minutes)   # default bandwidth
density = kde(x_grid)

axes[0].plot(x_grid, density)
axes[0].set_ylabel("Density")
#axes[0].set_xlabel("Minute within hour")
axes[0].set_title("KDE of Incident Times (Full Dataset)")

# KDE on 25 random points 
np.random.seed(42)
sample_25 = np.random.choice(minutes, size=25, replace=False)

kde_25 = gaussian_kde(sample_25)
density_25 = kde_25(x_grid)

axes[1].plot(x_grid, density_25)
axes[1].scatter(sample_25, np.zeros_like(sample_25), alpha=0.6)

#axes[1].set_xlabel("Minute within hour")
axes[1].set_ylabel("Density")
axes[1].set_title("KDE using only 25 Random Time Points")

print('KDEs can be misleading because they invent smooth structure even when data is sparse.')
````

````python
# Create a probability plot

from scipy.stats import probplot


fig, axes = plt.subplots(1, 2, figsize=(12, 5))

for ax, crime_i in zip(axes, crimes):
    X = df_filtered.loc[
        df_filtered['Incident Category'] == crime_i,
        'Longitude'
    ]
    X = pd.to_numeric(X, errors='coerce').dropna()

    probplot(X, dist="norm", plot=ax)
    ax.set_title(f"Normal Q–Q Plot\n{crime_i} (Longitude)")
    ax.set_xlabel("Theoretical quantiles")
    ax.set_ylabel("Observed latitude")

plt.tight_layout()
plt.show()

# What reference distribution are you comparing against? 
#   Normal bell-shaped distribution 
# What would it mean if the points fell exactly on the straight line? 
#   It would mean that crime locations are spread out in a smooth, bell-shaped way from north to south
# Why might the latitude distributi°on of crimes deviate from normal 
#   Some neighbourhoods might have a higher crime rate than others
# What does the deviation tell you about the geography of crime in SF?
#   It shows that crime is clustered in certain neighborhoods, not spread evenly across the city.
````

````python
# Calculate the average for each focus-crime, namely what time of day the crime happens. 
# So this time, the distribution we want to plot is the average time-of-day that a crime takes place. 
# For datapoint, the only thing you care about is the time-of-day, so discard everything else.
    # Convert it to hour and minute/second as a percentage of an hour
    
dt = pd.to_datetime(df['Incident Datetime'])

df_time_of_day = pd.DataFrame({
    'time_of_day': dt.dt.hour + (dt.dt.minute * 60 + dt.dt.second)/ 3600.0,
    'datetime': dt,
    'Incident Category': df['Incident Category']
})


data = [
    df_time_of_day.loc[df_time_of_day['Incident Category'] == crime, 'time_of_day']
                  .dropna()
                  .to_numpy()
    for crime in focus_crimes
]

fig, ax = plt.subplots(figsize=(10, 7))
bp = ax.boxplot(data, labels=focus_crimes)

# y-axis ticks every 3 hours
yticks = np.arange(0, 25, 3)
ax.set_yticks(yticks)

# map shifted time back to clock time
ylabels = [f"{int((y) % 24):02d}:00" for y in yticks]
ax.set_yticklabels(ylabels)

ax.set_ylabel("Incidents per day")
ax.set_title("Distribution of daily incident counts by category")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
plt.show()

# For crimes that peak late at night, does the box plot do a good job of showing the "typical" time?
# What goes wrong when the distribution wraps around midnight (e.g., a crime that peaks between 11pm and 1am)? 
# Can you think of a way to handle this?
#   Its weird that the peak is around 14/15 in all cases, I will try to shift it around for 6 hours and see what happens

df_time_of_day_shift = pd.DataFrame({
    'time_of_day': (df_time_of_day['time_of_day'] - 6) % 24,
    'Incident Category': df_time_of_day['Incident Category']
})


data = [
    df_time_of_day_shift.loc[df_time_of_day_shift['Incident Category'] == crime, 'time_of_day']
                  .dropna()
                  .to_numpy()
    for crime in focus_crimes
]

fig, ax = plt.subplots(figsize=(10, 7))
bp = ax.boxplot(data, labels=focus_crimes)

ax.set_ylabel("Incidents per day")
ax.set_title("Distribution of daily incident counts by category shifted around 6 hours")
plt.xticks(rotation=45, ha='right')

# y-axis ticks every 3 hours
yticks = np.arange(0, 25, 3)
ax.set_yticks(yticks)

# map shifted time back to clock time
ylabels = [f"{int((y + 6) % 24):02d}:00" for y in yticks]
ax.set_yticklabels(ylabels)

plt.tight_layout()
plt.show()
plt.show()

# THIS ACTUALLY DOESNT WORK, IT LOOKS EXACTLY THE SAME AFTER THE SHIFT ://
# THIS METHOD ACTUALLY DOESNT MAKE SENSE BECAUSE LATE HOURS HAVE MORE VALUE THEN EARLIER HOURS
````
